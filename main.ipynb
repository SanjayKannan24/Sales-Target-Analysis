##### Sales 2017

```python
sales2017 = pd.read_excel("Source/Sales 2017.xlsx")
sales2017.head()
```

```python
sales2017.columns
```

```python
# Handling missing values
missingData = sales2017.isnull().sum()

plt.figure(figsize=(16, 4))
sns.pointplot(x=missingData.index, y=missingData.values)
plt.xlabel("Columns")
plt.ylabel("Percentage of Missing data")
plt.xticks(rotation= 90)
plt.show()
```

```python
sales2017.count()
```

```python
sales2017["Product ID"].nunique()
```

```python
sales2017["Customer ID"].nunique()
```

```python
sales2017["Total"] = sales2017["Qty Itens"] * sales2017["Unit Price"]
```

```python
sales2017.head()
```

```python
df2017 = pd.merge(sales2017, dates, left_on='Issue Date', right_on='Dates')
```

```python
df2017.head()
```

```python
df2017 = pd.merge(df2017, product, on='Product ID')
df2017.head()
```

```python
df2017 = pd.merge(df2017, productGroup, on="Group ID")
df2017.head()
```

```python
df2017 = pd.merge(df2017, salesPerson, on='SalesPerson ID')
df2017 = pd.merge(df2017, customer, on='Customer ID')
df2017.head()
```

```python
df2017 = df2017.drop(['Issue Date','Due Date', 'Invoice Number', 'Net Weight', 'Supervisor ID', 'Manager ID', 'State'], axis=1)
```

```python
df2017 = df2017.drop(['Customer ID', 'Month', 'Year', 'Product ID'], axis=1)
```

```python
df2017 = df2017.drop(['Group ID', 'Day'], axis=1)
df2017.head()
```

```python
# Group the merged DataFrame by "Month Name" and calculate the sum of "Total" for each month
monthly_totals = df2017.groupby('Month Name')['Total'].sum().reset_index()

month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

# Create the countplot
plt.figure(figsize=(10, 6))  # Set the figure size (adjust as needed)
ax = sns.barplot(x='Month Name', y='Total', data=monthly_totals, order=month_order)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

plt.title('Total by Month Name')

for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline')

plt.show()
```

```python
# plt.pie(df2017.groupby('Category')['Group'].sum().index)

pie_graph = df2017.groupby('Category')['Group'].nunique()


plt.figure(figsize=(4, 4))
plt.pie(pie_graph, labels=pie_graph, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.

plt.title('Distribution of Groups within Categories')
plt.show()
```

```python
ManagerWiseTotal = df2017.groupby(['Manager', 'Team'])['Total'].sum().reset_index()

plt.figure(figsize=(10, 5))
ax = sns.barplot(x='Manager', y='Total', hue='Team', data= ManagerWiseTotal)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline')
```

```python
LocationWiseTotal = df2017.groupby(['City'])['Total'].sum().reset_index()

LocationWiseTotal = LocationWiseTotal.sort_values(by='Total', ascending=False)

plt.figure(figsize=(16, 5))
ax = sns.barplot(x='City', y='Total', data= LocationWiseTotal)
plt.xticks(rotation=45)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline')
```
